```python
# Importar las bibliotecas
import pandas as pd
import polars as pl

# Crear un DataFrame con Pandas
data_pandas = {
    "nombre": ["Ana", "Luis", "María", "Juan"],
    "edad": [23, 45, 34, 25],
    "ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla"]
}
df_pandas = pd.DataFrame(data_pandas)

# Operaciones básicas con Pandas
print("DataFrame con Pandas:")
print(df_pandas)
print("\nSeleccionar la columna 'nombre':")
print(df_pandas["nombre"])
print("\nFiltrar filas donde la edad es mayor de 30:")
print(df_pandas[df_pandas["edad"] > 30])

# Crear un DataFrame con Polars
df_polars = pl.DataFrame(data_pandas)

# Operaciones básicas con Polars
print("\nDataFrame con Polars:")
print(df_polars)
print("\nSeleccionar la columna 'nombre':")
print(df_polars.select("nombre"))
print("\nFiltrar filas donde la edad es mayor de 30:")
print(df_polars.filter(pl.col("edad") > 30))
```
