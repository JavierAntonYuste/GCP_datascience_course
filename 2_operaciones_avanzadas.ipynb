```python
# Importar las bibliotecas
import pandas as pd
import polars as pl

# Crear un DataFrame con Pandas
data_pandas = {
    "nombre": ["Ana", "Luis", "María", "Juan", "Lucía"],
    "edad": [23, 45, None, 25, 30],
    "ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla", "Bilbao"],
    "salario": [50000, 60000, 70000, 80000, 90000]
}
df_pandas = pd.DataFrame(data_pandas)

# Operaciones avanzadas con Pandas
print("DataFrame con Pandas:")
print(df_pandas)
print("\nRellenar datos faltantes en la columna 'edad':")
df_pandas["edad"].fillna(df_pandas["edad"].mean(), inplace=True)
print(df_pandas)
print("\nAgrupar por 'ciudad' y calcular la media del salario:")
print(df_pandas.groupby("ciudad")["salario"].mean())

# Crear un DataFrame con Polars
df_polars = pl.DataFrame(data_pandas)

# Operaciones avanzadas con Polars
print("\nDataFrame con Polars:")
print(df_polars)
print("\nRellenar datos faltantes en la columna 'edad':")
df_polars = df_polars.with_columns(
    pl.col("edad").fill_nan(df_polars["edad"].mean())
)
print(df_polars)
print("\nAgrupar por 'ciudad' y calcular la media del salario:")
print(df_polars.group_by("ciudad").agg(pl.col("salario").mean()))
```
